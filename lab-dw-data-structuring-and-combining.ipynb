```python
import pandas as pd
```

```python
url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df1 = pd.read_csv(url1)
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
df2 = pd.read_csv(url2)
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
df3 = pd.read_csv(url3)
```

```python
df1.columns
```

```python
df2.columns
```

```python
df3.columns
```

```python
df1.columns = df1.columns.str.lower().str.replace(" ", "_")
df1.columns
```

```python
df1.columns = ['customer', 'state', 'gender', 'education', 'customer_lifetime_value', 'income', 'monthly_premium_auto', 'number_of_open_complaints', 'policy_type', 'vehicle_class', 'total_claim_amount']
df1.columns
```

```python
df2.columns = df2.columns.str.lower().str.replace(" ", "_")
df2.columns
```

```python
df2.columns = ['customer', 'state', 'gender', 'education', 'customer_lifetime_value', 'income', 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount', 'policy_type', 'vehicle_class']
df2.columns
```

```python
df3.columns = df3.columns.str.lower().str.replace(" ", "_")
df3.columns
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
print(df1.isnull().sum())
print(df2.isnull().sum())
```

```python
print(df3.isnull().sum())
```

```python
df1.sample(20)
```

```python
combined_df = pd.concat([df1, df2, df3], axis=0)
combined_df
```

```python
combined_df.isnull().sum()
```

```python
combined_df.shape
```

```python
combined_df = combined_df.dropna(how = "all")
```

```python
combined_df.isnull().sum()
```

```python
combined_df.duplicated().sum()
```

```python
combined_df = combined_df.drop_duplicates()
```

```python
combined_df.duplicated().sum()
```

```python
combined_df.reset_index(drop=True, inplace=True)
```

```python
combined_df.duplicated(subset = ["customer"]).sum()
```

```python
combined_df = combined_df.drop_duplicates(subset=["customer"])
```

```python
combined_df.duplicated(subset = ["customer"]).sum()
```

```python
combined_df.isnull().sum()
```

```python
combined_df["gender"] = combined_df["gender"].fillna("Unknown")
```

```python
combined_df.isnull().sum()
```

```python
combined_df[combined_df["customer_lifetime_value"].isna()]
```

```python
combined_df["customer_lifetime_value"] = pd.to_numeric(combined_df["customer_lifetime_value"].str.replace("%", "", regex=False), errors="coerce")
```

```python
combined_df["customer_lifetime_value"].head()
```

```python
combined_df["customer_lifetime_value"].mean()
```

```python
combined_df["customer_lifetime_value"].median()
```

```python
combined_df["customer_lifetime_value"] = combined_df["customer_lifetime_value"].fillna(combined_df["customer_lifetime_value"].median())
```

```python
combined_df.isnull().sum()
```

```python
combined_df
```

```python
combined_df = combined_df.reset_index()
combined_df
```

```python
combined_df = combined_df.drop(columns=["level_0", "index"])
```

```python
combined_df
```

```python
combined_df["gender"].unique()
```

```python
combined_df["gender"] = combined_df["gender"].str.strip().str.lower().replace({"female": "F", "femal": "F", "f": "F", "male": "M", "m": "M"})
combined_df["gender"].unique()
```

```python
combined_df["state"].unique()
```

```python
combined_df["state"] = combined_df["state"].replace({"AZ": "Arizona", "WA": "Washington", "Cali": "California"})
```

```python
combined_df["state"].unique()
```

```python
combined_df["education"].unique()
```

```python
combined_df["education"] = combined_df["education"].replace({"Bachelors": "Bachelor"})
```

```python
combined_df["education"].unique()
```

```python
combined_df["vehicle_class"] = combined_df["vehicle_class"].replace({"Luxury SUV": "Luxury", "Luxury Car": "Luxury", "Sports Car": "Luxury"})
```

```python
combined_df["vehicle_class"].unique()
```

```python
combined_df
```

```python
combined_df.isnull().sum()
```

```python
def clean_complaints(value):
    value = str(value)
    
    if "/" in value:
        return value.split("/")[1]
    else:
        return value

combined_df["number_of_open_complaints"] = (
    combined_df["number_of_open_complaints"]
    .apply(clean_complaints)
)

combined_df["number_of_open_complaints"] = pd.to_numeric(
    combined_df["number_of_open_complaints"],
    errors="coerce"
)
```

```python
combined_df["number_of_open_complaints"].unique()
```

```python
combined_df
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
df_mkt["sales_channel"].unique()
```

```python
pivot_df = pd.pivot_table(df_mkt, index="sales_channel", values="total_claim_amount", aggfunc="sum").round(2)
pivot_df
```

```python
pivot_df["Percentage"] = (pivot_df["total_claim_amount"] / pivot_df["total_claim_amount"].sum() * 100).round(2)
pivot_df
```

Firstly, I would like to note that the exercise asks for mail as a sales channel. However, when checking the options in sales channel, mail is not one. The channels used were agent, branch, call center and web.

By looking at the pivot table we can easily see that the agent channel was the most effective, 38%, followed not by far by branch.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_lifetime = pd.pivot_table(df_mkt, index="gender", columns="education", values=["customer_lifetime_value"], aggfunc="mean").round(2)
pivot_lifetime
```

The pivot shows the average customer lifetime value (CLV) varies slightly across education and gender. Females with High School or below have the highest clv, 8675.22, while male customer with a Master have the highest clv , 8168.83. Difference between gender is small, which means that gender does not impact a lot on clv.
